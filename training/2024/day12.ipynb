```python

from collections import deque

def solve_p1(inp:str) -> float:
    grid= np.array([list(line.strip()) for line in inp.split("\n") if line.strip()], dtype=str)
    

    price=0
    nb_rows, nb_cols = grid.shape   
   
    def fill(row, col, grid:np.ndarray) -> tuple[float, set]:
        """area * perimeter, if not seen_coords else 0.0"""
        nb_rows, nb_cols = grid.shape
        type_ = grid[row, col]
        seen_coords=set()
        curr_area = curr_perim = 0
        queue = deque([(row, col)])
        while queue:
            row, col = queue.popleft()
            if any([row<0, col<0, row>=nb_rows, col>=nb_cols]): # outside the map
                curr_perim+=1
                seen_coords.add((row, col))  
                continue
            curr_type=grid[row, col]
            if (row, col) in seen_coords: continue
            
            if curr_type != type_: # crossed border into another region
                curr_perim+=1
                continue
            curr_area+=1
            
            seen_coords.add((row, col))            
            queue.extend([(row-1,col), (row+1, col), (row, col-1), (row, col+1)])
            
        print(f"{curr_area=},{curr_perim=}")
        return curr_area*curr_perim, seen_coords
    
    seen=set()
    for r in range(nb_rows):
        for c in range(nb_cols):
            if (r,c) in seen: continue
            add_price, new_seen = fill(r, c, grid)
            price+=add_price
            seen |=new_seen
            
    
    print(price)
    return price
    
```
